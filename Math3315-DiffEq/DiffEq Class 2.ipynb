```python
def f(t,u):
  r=5
  H=5
  return r*u*(1-u/H)

N=100
T=5
h=T/N
t=np.linspace(0,T,N+1)
u=np.empty(N+1)
u[0]=0.1

for k in range(N):
  u[k+1]=u[k]+h*f(t[k],u[k])
```

```python
import matplotlib.pyplot as plt

plt.plot(t,u)
plt.show()
```
